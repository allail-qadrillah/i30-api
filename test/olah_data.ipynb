```python
def print_all_unique_col_values(df):
  for column in df:
      print(f"{column} : {df[column].unique()}")


def print_object_unique_col_values(df):
  for column in df:
    if df[column].dtypes == 'object':
      print(f"{column} : {df[column].unique()}")


print_object_unique_col_values(df2)
```
